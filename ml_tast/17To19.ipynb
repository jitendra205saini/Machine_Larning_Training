```python
import numpy as np  # Yeh library numerical calculations ke liye hai
import pandas as pd  # Yeh library data ko table format mein handle karne ke liye hai
import matplotlib.pyplot as plt  # Yeh library charts banane ke liye hai
```

# QUES. 17

```python
df = pd.read_csv(r'D:\data_science\ml_test\Book1.csv')         # Ye line CSV file ko read karta hai jo is path pe hai
salaries = df['Salary'].to_numpy()                             # Ye 'Salary' column ko leke usko NumPy array mein convert karta hai
average_salary = np.mean(salaries)                             # Ye salaries ka average nikalne ke liye NumPy ka mean function use karta hai
print("Average Salary:", average_salary)                       # Ye average salary ko print karta hai taki hum dekh sake result
```

# QUES. 18

```python
#  Filter out employees with salary greater than 60000
filtered_df = df[df['Salary'] > 60000] 
print(filtered_df)
```

# QUES. 19

```python
# Yeh line department ke hisaab se salary ka average nikaalti hai aur sort karti hai alphabetically
dept_avg_salary = df.groupby('Department')['Salary'].mean().sort_index()

# Yeh line bar chart banati hai, x-axis pe departments aur y-axis pe avg salary
plt.bar(dept_avg_salary.index, dept_avg_salary.values)

plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Department-wise Average Salary')
plt.show()
```
