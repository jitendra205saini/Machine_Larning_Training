```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset from CSV file
file_path = "D:/data_science/ml_test/department_salary.csv"  
df = pd.read_csv(file_path)
```

# QES. 20

```python
#  Bar Chart - Number of employees in each department
plt.figure(figsize=(6, 4))
df["Department"].value_counts().plot(kind="bar", color=["blue", "green", "red"])
plt.title("Number of Employees in Each Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

# QES. 21

```python
#  Filtering employees with salary greater than 60,000
high_salary_df = df[df["Salary"] > 60000]
print("Employees with Salary > 60,000:\n", high_salary_df)


```

# QES. 22

```python
#  Line Graph - Salary trend by department
plt.figure(figsize=(6, 4))
for dept in df["Department"].unique():
    subset = df[df["Department"] == dept]
    plt.plot(subset["Name"], subset["Salary"], marker="o", label=dept)

plt.title("Salary Trend by Department")
plt.xlabel("Employee")
plt.ylabel("Salary")
plt.legend(title="Department")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
