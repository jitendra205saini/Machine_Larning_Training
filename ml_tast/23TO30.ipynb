```python
import numpy as np  # Yeh library numerical calculations ke liye hai
import pandas as pd  # Yeh library data ko table format mein handle karne ke liye hai
import matplotlib.pyplot as plt  # Yeh library charts banane ke liye hai
```

```python
# CSV file read karein
df = pd.read_csv(r"D:\data_science\ml_test\Screenshot 2025-03-28 072707.csv")  
```

# QUES. 23

```python
# Har subject ka mean score calculate karna
subject_means = df.iloc[:, 1:].mean()  # Pehla column chhodkar baaki columns ka mean nikalta hai
print("Mean Score of Each Subject:")
print(subject_means)
```

# QUES. 24

```python
# Har student ka average score calculate karna
df["Average Score"] = df.iloc[:, 1:].mean(axis=1)  # Pehla column chhodkar baaki ka row-wise mean
print("\nAverage Score of Each Student:")
print(df[["Student Name", "Average Score"]])
```

# QUES. 25

```python
# DataFrame me se us student ka naam nikalne ke liye jo highest overall score rakhta hai
top_student = df.loc[df["Average Score"].idxmax(), "Student Name"]
print("\nStudent with Highest Overall Score:", top_student)
```

# QUES. 26

```python

# Math aur Science ke scores ka scatter plot banane ke liye
plt.figure(figsize=(6, 4)) # graph size set karta hai
plt.scatter(df["Math"], df["Science"], color='blue') # Scatter plot draw kar raha hai Math vs Science scores ke liye
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.title("Math vs Science Scores")
plt.show()

```

# QUES. 27   

```python
# Scores ko standardize karne ke liye (mean 0 aur std deviation 1 banane ke liye)
standardized_scores = (df.iloc[:, 1:-1] - df.iloc[:, 1:-1].mean()) / df.iloc[:, 1:-1].std()
print("\nStandardized Scores:")
print(standardized_scores)
```

# QUES. 28

```python
# Har student ka total score calculate kar raha hai
df["Total Score"] = df.iloc[:, 1:].sum(axis=1)
plt.figure(figsize=(6, 4))
plt.bar(df["Student Name"], df["Total Score"], color='green') # Bar plot bana raha hai (X-axis: Student Name, Y-axis: Total Score)
plt.xlabel("Student Name")
plt.ylabel("Total Score")
plt.title("Total Scores of Students")
plt.show()
```

# QUES. 30

```python
avg_history_score = df["History"].mean() # History subject ka average score nikal raha hai
high_history_students = df[df["History"] > avg_history_score] # Jo students ka History score average se zyada hai unko filter kar raha hai
print("\nStudents with History Score greater than Average:")
print(high_history_students[["Student Name", "History"]])
```
