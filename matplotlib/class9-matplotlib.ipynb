# 4. Matplotlib Labels and Title

### Create Labels for a Plot

```python
xpoint = np.array([3,8,1,10])
ypoint = np.array([3,6,4,10])
plt.plot(xpoint,ypoint)
plt.xlabel('sumit')
plt.ylabel('karuna')


plt.show()
```

### Create a Title for a Plot

```python
xpoint = np.array([3,8,1,10])
ypoint = np.array([3,6,4,10])

plt.plot(xpoint,ypoint)

plt.title('graph')
plt.xlabel('sumit')
plt.ylabel('karuna')

plt.show()
```

### Set Font Properties for Title and Labels

```python
xpoint = np.array([3,8,1,10])
ypoint = np.array([3,6,4,10])

f1 = {'family':'serif','color':'blue','size':20}
f2 = {'family':'serif','color':'darkred','size':15}

plt.title('graph',fontdict=f2)
plt.xlabel('sumit',fontdict=f1)
plt.ylabel('karuna',fontdict=f1)

plt.plot(xpoint,ypoint)
plt.show()
```

### Position the Title

```python
xpoint = np.array([3,8,1,10])
ypoint = np.array([3,6,4,10])

plt.plot(xpoint,ypoint)

plt.title('graph',loc='left')
plt.xlabel('sumit')
plt.ylabel('karuna')

plt.show()
```

# 5. Matplotlib Adding Grid Lines

### Add Grid Lines to a Plot

```python
xpoint = np.array([3,8,1,10])
ypoint = np.array([3,6,4,10])

plt.title('graph')
plt.xlabel('sumit')
plt.ylabel('karuna')

plt.plot(xpoint,ypoint)

plt.grid()
plt.show()
```

### Specify Which Grid Lines to Display

```python
xpoint = np.array([3, 8, 1, 10])
ypoint = np.array([3, 6, 4, 10])

plt.title('graph')
plt.xlabel('sumit')
plt.ylabel('karuna')

plt.plot(xpoint, ypoint)

plt.grid(axis='x') # Sirf x-axis ke liye grid
plt.show()
```

```python
xpoint = np.array([3, 8, 1, 10])
ypoint = np.array([3, 6, 4, 10])

plt.title('graph')
plt.xlabel('sumit')
plt.ylabel('karuna')

plt.plot(xpoint, ypoint)

plt.grid(axis='y') # Sirf y-axis ke liye grid
plt.show()
```

### Set Line Properties for the Grid

```python
xpoint = np.array([3, 8, 1, 10])
ypoint = np.array([3, 6, 4, 10])

plt.title('graph')
plt.xlabel('sumit')
plt.ylabel('karuna')

plt.plot(xpoint, ypoint)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

### Size

```python
x = np.array([1,2,3,4])
y = np.array([3,7,1,8])
s= np.array([330,5,860,170])
plt.scatter(x,y,s = s)
plt.show()
```

### Alpha

```python
x = np.array([1,2,3,4])
y = np.array([3,7,1,8])
s= np.array([330,500,860,170])
c = np.array(['red','green','blue','black'])
plt.scatter(x,y,s = s,c =c ,alpha = 0.5) #You can adjust the transparency of the dots
plt.show()
```

### Combine Color Size and Alpha

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# 7. Matplotlib Bars

### Explode

```python
x = np.array([5,5,5,5])

mylabel = ["red","green","blue","black"]
myexplode =[0.1,0,0,0]
plt.pie(x,labels = mylabel,explode=myexplode,startangle= 90)

plt.show()
```

### Shadow

```python
x = np.array([5,5,5,5])

mylabel = ["red","green","blue","yellow"]
myexplode =[0.1,0,0,0]

plt.pie(x,labels = mylabel,explode=myexplode,startangle= 90,shadow= True)


plt.show()
```

### Colors

```python
x = np.array([5,5,5,5])

mylabel = ["red","green","blue","yellow"]
myexplode =[0.1,0,0,0]
mycolors = ["red","green","blue","yellow"]
plt.pie(x,labels = mylabel,explode=myexplode,startangle= 90,shadow= True,colors =mycolors)


plt.show()
```

### Legend

```python
x = np.array([5,5,5,5])

mylabel = ["red","green","blue","yellow"]
myexplode =[0.1,0,0,0]
mycolors = ["red","green","blue","yellow"]
plt.pie(x,labels = mylabel,explode=myexplode,startangle= 90,shadow= True,colors =mycolors)
plt.legend()

plt.show()
```

### Legend With Header

```python
x = np.array([5,5,5,5])

mylabel = ["red","green","blue","yellow"]
myexplode =[0.1,0,0,0]
mycolors = ["red","green","blue","yellow"]
plt.pie(x,labels = mylabel,explode=myexplode,startangle= 90,shadow= True,colors =mycolors)
plt.legend(title = "color")

plt.show()
```

### Yeh autopct ke kaam ko focus karta hai (percentage dikhana).

```python
x = np.array([5,5,5,5])

mylabel = ["red","green","blue","yellow"]
myexplode =[0.1,0,0,0]
mycolors = ["red","green","blue","yellow"]
plt.pie(x,labels = mylabel,explode=myexplode,startangle= 90,shadow= True,colors =mycolors,autopct ='%1.1f%%')
plt.legend(title = "color")

plt.show()
```
